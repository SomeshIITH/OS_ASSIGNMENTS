```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
chunk_time = [207.6,267.4,317.6,1245.8,2975.2,13296.8,22885.6,38244.2]
mixed_time = [256.8,244.8,290.8,1256,2936,13577.8,21583.2,37684.6]
seq_time = [117,195,594.4,4193.8,9675.6,46687.4,91000.8,126008]
dim = [25,49,100,625,1024,2500,3136,4096]

plt.figure(figsize=(15,10))

plt.plot(dim,chunk_time,'-o',label="Chunk")
plt.plot(dim,mixed_time,'-',label = "Mixed")
plt.plot(dim,seq_time,'--',label = "Sequential")

plt.xlabel('Dimension of Matrix')
plt.ylabel('Time taken in microseconds')
plt.title('Time taken Vs Dimension (Taking K =8 threads)')

plt.grid()
plt.legend()
plt.show()
```
